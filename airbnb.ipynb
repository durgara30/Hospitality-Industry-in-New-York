```python
#checking for outliers

plt.figure(figsize=[10,10])
ab=sns.boxplot(data=airbnb,x='room_type',y='price',palette='plasma')
```

```python
#removing outliers

def get_outliers(tempdf):
    Q1=tempdf.price.quantile(0.25)
    Q3=tempdf.price.quantile(0.75)
    IQR=Q3-Q1
    
    upper=Q3+1.5*IQR
    upper_array=np.array(tempdf[tempdf['price']>=upper].index)
    
    lower=Q1-1.5*IQR
    lower_array=np.array(tempdf[tempdf['price']<=lower].index)
    
    removable_outlier_locations=np.append(upper_array,lower_array)
    return removable_outlier_locations

locations_to_be_removed=np.array([])

for boroughs in list(airbnb['borough'].unique()):
    for roomtypes in list(airbnb['room_type'].unique()):
        tempdf=airbnb[(airbnb.borough==boroughs)&(airbnb.room_type==roomtypes)]
        locations_to_be_removed=np.append(locations_to_be_removed,get_outliers(tempdf))
        
    print(str(len(locations_to_be_removed))+' outlier removed!')
    airbnb_without_borough_outliers=airbnb.drop(locations_to_be_removed)
print('Output saved in airbnb_without_borough_outliers')
```

```python
#saving the outliers removed dataframe into excel file

airbnb_without_borough_outliers.to_excel('Airbnb_cleaned.xlsx')
```
